## Merging dataframes

### Merge two dataframes [inner, left, right, outer join]

Use students_list.csv files

````python
import pandas as pd
import numpy as np
````

````python
df = pd.read_csv("../csv-files/data_studentlist.csv")
````

````python
df2.shape
````

#### Inner Join

Returns only the rows that have matching values in both DataFrames.

````python
df1.merge(right=df2, how="inner", on="name")
````

#### Left Join

Returns all rows from the left DataFrame, and the matched rows from the right.

````python
df1.merge(df2, on="name", how="left")
````

#### Right Join

Returns all rows from the right DataFrame, and the matched rows from the left DataFrame.

````python
df1.merge(df2, on="name", how="right")
````

#### Outer Join

Returns all rows when there is a match in either left or right DataFrame.

````python
df1.merge(df2, on="name", how="outer")
````

## Concatenation

Concatenating means appending data from different sources into a single DataFrame.

### Merge data from `data.csv` and `data.json` files
- How can we concat concat function

````python
df_csv = pd.read_csv("../csv-files/data.csv")
df_csv = df_csv.iloc[:, 1:]
df_csv.head()
````

````python
df_json = pd.read_json("../csv-files/data.json")
df_json
````

````python
df_csv.shape, df_json.shape
````

````python
df_csv.columns, df_json.columns
````

````python
df_csv["mpg"] = df_csv["mpg"] * 0.425144
df_csv.rename(columns={"mpg": "kmpl"}, inplace=True)
df_csv.head()
````

````python
df_csv["ground_clearance_mm"] = df_csv["ground_clearance_mm"] / 10
````

````python
df_csv.rename(columns={"horsepower": "hp", "name": "Name", "ground_clearance_mm": "ground_clearance", "origin": "country"}, inplace=True)
df_csv.head()
````

````python
concat_data = pd.concat([df_csv, df_json], ignore_index=True, sort=False)
print("Concatenated Data:")
print(concat_data.head())
````

````python
concat_data.shape
````

### What are the similarities and difference in the two sources?

**Similarities:** Both sources contain automotive data with common columns such as 'cylinders' and 'displacement'.

**Differences:** The CSV source includes 'mpg', 'horsepower', 'weight', etc., while the JSON source has 'hp', 'kmpl', and 'ground_clearance'. We had to convert the units, rename the mismatching columns as well.

## Using Axes

````python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
````

### Multiple plots in the same axes

````python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # [left, bottom, width, height]

ax.plot(x, y1, label='Sine')
ax.plot(x, y2, label='Cosine')
ax.legend()
ax.set_title('Sine and Cosine Waves')

plt.show()
````

### Multiple plots in different axes

````python
fig = plt.figure(figsize=(12, 6))

ax1 = fig.add_axes([0.1, 0.1, 0.35, 0.8])
ax1.plot(x, y1, label='Sine', color='blue')
ax1.legend()
ax1.set_title('Sine Wave')

ax2 = fig.add_axes([0.55, 0.1, 0.35, 0.8])
ax2.plot(x, y2, label='Cosine', color='green')
ax2.legend()
ax2.set_title('Cosine Wave')

plt.show()
````

### Multiple plots using plt.subplots

````python
y3 = np.tan(x)
y4 = np.exp(x)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Sine
axs[0, 0].plot(x, y1, label='Sine', color='blue')
axs[0, 0].legend()
axs[0, 0].set_title('Sine Wave')

# Plot 2: Cosine
axs[0, 1].plot(x, y2, label='Cosine', color='green')
axs[0, 1].legend()
axs[0, 1].set_title('Cosine Wave')

# Plot 3: Tangent
axs[1, 0].plot(x, y3, label='Tangent', color='red')
axs[1, 0].legend()
axs[1, 0].set_title('Tangent Wave')

# Plot 4: Exponential
axs[1, 1].plot(x, y4, label='Exponential', color='purple')
axs[1, 1].legend()
axs[1, 1].set_title('Exponential Wave')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
````

### Feature selection

````python
df = pd.read_csv("../csv-files/data_studentlist.csv")


# Select the features and the target variable
X = df[['age', 'grade', 'height', 'weight']]
y = df['grade']  # Assuming 'grade' is the target variable for regression

# Create SelectKBest object with f_regression
k = 3  # Select the top 3 features, you can change this as needed
selector = SelectKBest(score_func=f_regression, k=k)

# Fit and transform the features
X_new = selector.fit_transform(X, y)

# Get the selected feature names
selected_features = X.columns[selector.get_support()]

# Create a DataFrame with the selected features
selected_features_df = pd.DataFrame(X_new, columns=selected_features)

# Combine the selected features with the original dataset
df_selected = pd.concat([df[['name', 'gender', 'age', 'grade', 'bloodtype', 'height', 'weight']], selected_features_df], axis=1)

print("Student List with Selected Features:")
print(df_selected.head())
````

````python
features = df[['age', 'grade', 'height', 'weight', 'BMI']]

poly = PolynomialFeatures(degree=2, include_bias=False)

# Fit and transform the features
poly_features = poly.fit_transform(features)

# Create a DataFrame with the polynomial features
poly_features_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(features.columns))

# Combine the polynomial features with the original dataset
df_poly = pd.concat([df, poly_features_df], axis=1)

print("Student List with Polynomial Features:")
print(df_poly.head())
````

In the above `data`, the Temperature column contains inconsistent capitalization ("Warm" vs "warm" and "Hot" vs "hot"). This inconsistency can cause issues during encoding.

### One Hot Encoding

First let's create a sample dataset for the task.

````python
data = {
    "Rec-no": range(10),
    "Temperature": ["Hot", "Cold", "Very Hot", "Warm", "Hot", "Warm", "warm", "hot", "hot", "cold"],
    "Color": ["Red", "Yellow", "Blue", "Blue", "Red", "Yellow", "Red", "Yellow", "Yellow", "Yellow"],
    "Target": [1, 1, 1, 0, 1, 0, 1, 0, 1, 1]
}

df = pd.DataFrame(data)
````

````python

df['Temperature'] = df['Temperature'].str.capitalize()
print("Original Data:")
print(df)
````

#### One-Hot Encoding (OHE) using `pd.get_dummies`

````python
df_ohe = pd.get_dummies(df, columns=['Temperature', 'Color'], drop_first=True)
print("One-Hot Encoded Data (pd.get_dummies):")
print(df_ohe)
````

#### One-Hot Encoding using `OneHotEncoder` from `sklearn`

````python
from sklearn.preprocessing import OneHotEncoder

df['Temperature'] = df['Temperature'].astype('category')
df['Color'] = df['Color'].astype('category')

# One-Hot Encoding using OneHotEncoder from sklearn
ohe = OneHotEncoder(drop='first')
encoded_features = ohe.fit_transform(df[['Temperature', 'Color']])

# Creating a DataFrame from encoded features
encoded_df = pd.DataFrame(encoded_features.toarray(), columns=ohe.get_feature_names_out(['Temperature', 'Color']))

# Concatenating with original DataFrame
df_ohe_sklearn = pd.concat([df[['Rec-no', 'Target']], encoded_df], axis=1)
print("One-Hot Encoded Data (sklearn):")
print(df_ohe_sklearn)
````

### Label Encoding

````python
from sklearn.preprocessing import LabelEncoder

# Label Encoding
label_enc = LabelEncoder()
df['Temperature_Label'] = label_enc.fit_transform(df['Temperature'])
df['Color_Label'] = label_enc.fit_transform(df['Color'])
print("Label Encoded Data:")
print(df[['Rec-no', 'Temperature', 'Temperature_Label', 'Color', 'Color_Label', 'Target']])
````

### Ordinal Encoding

````python
from sklearn.preprocessing import OrdinalEncoder

# Ordinal Encoding
ord_enc = OrdinalEncoder()
df['Temperature_Ordinal'] = ord_enc.fit_transform(df[['Temperature']])
df['Color_Ordinal'] = ord_enc.fit_transform(df[['Color']])
print("Ordinal Encoded Data:")
print(df[['Rec-no', 'Temperature', 'Temperature_Ordinal', 'Color', 'Color_Ordinal', 'Target']])
````

### Frequency Encoding


````python
# Frequency Encoding
temp_freq = df['Temperature'].value_counts() / len(df)
color_freq = df['Color'].value_counts() / len(df)
df['Temperature_Freq'] = df['Temperature'].map(temp_freq)
df['Color_Freq'] = df['Color'].map(color_freq)
print("Frequency Encoded Data:")
print(df[['Rec-no', 'Temperature', 'Temperature_Freq', 'Color', 'Color_Freq', 'Target']])
````

#### Mean Encoding

````python

temp_mean = df.groupby('Temperature')['Target'].mean()
color_mean = df.groupby('Color')['Target'].mean()
df['Temperature_Mean'] = df['Temperature'].map(temp_mean)
df['Color_Mean'] = df['Color'].map(color_mean)
print("Mean Encoded Data:")
print(df[['Rec-no', 'Temperature', 'Temperature_Mean', 'Color', 'Color_Mean', 'Target']])
````

### Differences between

**fit_transform vs fit + transform:**

`fit_transform`: Fits the data and then transforms it in one step. It’s convenient when we want to apply both operations sequentially.

`fit` + `transform`: Fits the data using `fit` and then applies the transformation using `transform`. It is useful when we want to fit the model on one dataset and transform another dataset.

**transform vs predict:**
`transform`: Applies a transformation to the data based on what was learned during `fit`. It is used in preprocessing.
`predict`: Generates predictions from a fitted model. It is used in the context of making predictions with machine learning models.
