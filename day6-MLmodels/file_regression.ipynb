```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
model = LinearRegression()
```

```python
df = sns.load_dataset("mpg")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df["horsepower"] = df["horsepower"].fillna(df["horsepower"].mean())
```

```python
X = df[["cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year"]]
y = df["mpg"]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(pd.DataFrame(data=[[8, 307.0, 130.0, 3504, 12.0, 70]], columns=["cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year"]))
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(X_test, y_test)
```
